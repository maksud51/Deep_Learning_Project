```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import random
from PIL import Image
```

# Lungs

```python
parent_path = "/home/dev/projects/dicom_project/jpg_files/Lung"
image_list = [os.path.join(parent_path,image) for image in  os.listdir(parent_path)]
random.shuffle(image_list)
image_list = image_list[10:]
```

```python

rows = 2
cols = 5

fig = plt.figure(figsize=(12, 6))
for i in range(1, min(len(image_list), rows * cols) + 1):
    img_path = image_list[i - 1]
    img = cv2.imread(img_path)
    ax = fig.add_subplot(rows, cols, i)
    ax.imshow(img)
    ax.axis("off")
plt.tight_layout()

plt.show()
```

```python
seg_images_1 = []
for image_path in image_list:
    sample_image = cv2.imread(image_path)
    img = cv2.cvtColor(sample_image,cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(256,256))
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    _,thresh = cv2.threshold(gray, np.mean(gray), 255, cv2.THRESH_BINARY_INV)
    # plt.imshow(thresh)
    seg_images_1.append(thresh)
```

```python
seg_images_2 = []
for image_path in image_list:
    sample_image = cv2.imread(image_path)
    img = cv2.cvtColor(sample_image,cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(256,256))
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    _,thresh = cv2.threshold(gray, np.mean(gray), 255, cv2.THRESH_BINARY_INV)
    # plt.imshow(thresh)
    edges = cv2.dilate(cv2.Canny(thresh,0,255),None)
    seg_images_2.append(edges)
```

```python
columns = 10
fig, axes = plt.subplots(nrows=3, ncols=columns, figsize=(15, 6))
for i, img_path in enumerate(image_list[:columns]):
    img = cv2.imread(img_path)
    axes[0, i].imshow(img)
    axes[0, i].axis("off")

for i, seg_img in enumerate(seg_images_1[:columns]):
    image = Image.fromarray(seg_img)
    axes[1, i].imshow(image, cmap='viridis')
    axes[1, i].axis("off")


for i, seg_img in enumerate(seg_images_2[:columns]):
    image = Image.fromarray(seg_img)
    axes[2, i].imshow(image, cmap='viridis')
    axes[2, i].axis("off")

plt.tight_layout()
plt.show()
```
